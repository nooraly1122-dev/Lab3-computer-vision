```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()
img_path = next(iter(uploaded))
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
```

```python
min_val = np.min(img)
max_val = np.max(img)
stretched = ((img - min_val) / (max_val - min_val)) * 255
stretched = stretched.astype(np.uint8)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.title("Contrast Stretched")
plt.imshow(stretched, cmap='gray')
plt.axis('off')
plt.tight_layout()
plt.show()
# Step 4: Save result
cv2.imwrite("contrast_stretched.jpg", stretched)
files.download("contrast_stretched.jpg")
```

```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt
```

```python
print("Upload an image (grayscale or color):")
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python
img = cv2.imread(image_path)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gamma = 2.2 # You can try values like 0.4 (brighten) or >1 (darken)
gamma_corrected = np.array(255 * (img_gray / 255) ** gamma, dtype='uint8')
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected, cmap='gray')
plt.title(f"Gamma Corrected (γ={gamma})")
plt.axis('off')

31

plt.tight_layout()
plt.show()
```

```python
import cv2
from google.colab import files
import matplotlib.pyplot as plt
import numpy as np
```
